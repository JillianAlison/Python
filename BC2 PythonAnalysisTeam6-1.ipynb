```python
#import files 
test_file = 'test.csv'
test_og = pd.read_csv(test_file)
test = test_og.copy()
train_file = 'train.csv'
train_og = pd.read_csv(train_file)
train = train_og.copy()

#make list with id s for feature engineering 
train_id = train['Id']
test_id = test['Id']

#drop the sale price before feature engineering
train['SalePrice'] = np.log1p(train['SalePrice'])

#make list with sale price and features to be engineered 
train_label = train['SalePrice'].reset_index(drop = True)
train_features = train.drop(['SalePrice'], axis = 1)


#concat 
all_features = pd.concat([train_features, test]).reset_index(drop=True)

#make into strings to dummify and adjust categories 
all_features['MSSubClass'] = all_features['MSSubClass'].apply(str)
all_features['YrSold'] = all_features['YrSold'].astype(str)
all_features['MoSold'] = all_features['MoSold'].astype(str)
all_features['GarageQual'] = all_features['GarageQual'].fillna('No').map({'TA': 'Normal', 'Fa': 'Bad', 'Gd': 'Good', 'Po': 'Bad','Ex': 'Good'}).astype(str)
all_features['SaleType'] = all_features['SaleType'].fillna(all_features['SaleType'].mode()).map({'WD': 'Warranty', 'CWD': 'Warranty', 'VWD': "Warranty", 'New': "New", 'COD': "Other", 'Oth': "Other", 'Con': "Contract", 'ConLw': "Contract", 'ConLI': "Contract", 'ConLD': "Contract"}).astype(str)
all_features['Porch_Size'] = all_features['ScreenPorch'] + train['3SsnPorch'] + train['EnclosedPorch'] + train['OpenPorchSF'] + train['WoodDeckSF']
all_features['SaleType'] = all_features['SaleType'].fillna('No').map({'WD': 'Warranty', 'CWD': 'Warranty', 'VWD': "Warranty", 'New': "New", 'COD': "Other", 'Oth': "Other", 'Con': "Contract", 'ConLw': "Contract", 'ConLI': "Contract", 'ConLD': "Contract"}).astype(str)
all_features['GarageType'] = all_features['GarageType'].fillna('No').map({'Attchd': 'Attached', 'Detchd': 'Dettached', 'BuiltIn': 'Attached', 'CarPort': 'Outside','Basment': 'Attached', '2Types': 'Two'}).astype(str)
all_features['Functional'] = all_features['Functional'].fillna('Typ').map({'Typ': 'Normal', 'Min1': 'Small', 'Maj1': 'Big', 'Min2': 'Small','Mod': 'Medium', 'Maj2': 'Big', 'Sev': 'Big', 'Sal': 'Big'}).astype(str)
all_features['KitchenQual'] = all_features['KitchenQual'].fillna('TA').map({'TA': 'Normal', 'Fa': 'Bad', 'Gd': 'Good', 'Po': 'Bad','Ex': 'Good'}).astype(str)
all_features['Electrical'] = all_features['Electrical'].fillna('SBrkr').map({'SBrkr': 'SBrkr', 'FuseF': 'Fuse', 'FuseA': 'Fuse', 'FuseP': 'Fuse', 'Mix': 'Mix'}).astype(str)
all_features['HeatingQC'] = all_features['HeatingQC'].fillna('No').map({'TA': 'Normal', 'Fa': 'Bad', 'Gd': 'Good', 'Po': 'Bad','Ex': 'Good'}).astype(str)
all_features['Heating'] = all_features['Heating'].fillna('No').map({'GasA': 'Gas', 'GasW': 'Gas', 'Grav': 'Furnace', 'Wall': 'Furnace', 'OthW': 'Other', 'Floor': 'Furnace'}).astype(str)
all_features['BsmtFinType1'] = all_features['BsmtFinType2'].fillna('No').map({'GLQ': 'Good', 'ALQ': 'Medium', 'Unf': 'Bad', 'Rec': 'Medium', 'BLQ': 'Bad','LwQ': 'Bad'}).astype(str)
all_features['BsmtExposure'] = all_features['BsmtExposure'].fillna('No').astype(str)
all_features['BsmtCond'] = all_features['BsmtCond'].fillna('No').map({'TA': 'Normal', 'Fa': 'Bad', 'Gd': 'Good', 'Po': 'Bad','Ex': 'Good'}).astype(str)
all_features['ExterQual'] = all_features['ExterQual'].fillna('No').map({'TA': 'Normal', 'Fa': 'Bad', 'Gd': 'Good', 'Po': 'Bad','Ex': 'Good'}).astype(str)
all_features['MasVnrType'] = all_features['MasVnrType'].fillna(train['MasVnrType'].mode()[0])
all_features['Condition1'] = all_features['Condition1'].fillna('No').map({'Norm': 'Normal', 'Feedr': 'Adj', 'PosN': 'Near', 'Artery': 'Adj', 'RRAe': 'Close', 'RRNn': 'Close', 'RRAn': 'Adj', 'PosA': 'Adj', 'RRNe': 'Close'}).astype(str)
all_features['LotConfig'] = all_features['LotConfig'].fillna('No').map({'Inside': 'Inside', 'FR2': 'Front', 'Corner': 'Corner', 'CulDSac': 'CulDSac', 'FR3': 'Front'}).astype(str)
all_features['LotShape'] = all_features['LotShape'].fillna('No').map({'Reg': 'Reg', 'IR1': 'Ir', 'IR2': 'Ir', 'IR3': 'Ir'}).astype(str)
all_features['HasMiscVal'] = all_features['MiscVal'].map(lambda x: 1 if x > 0 else 0)
all_features['HasFireplace'] = all_features['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)
all_features['HasGarage'] = all_features['GarageCars'].apply(lambda x: 1 if x > 0 else 0)
all_features['Has2ndFloor'] = all_features['2ndFlrSF'].apply(lambda x: 1 if x > 0 else 0)
all_features['HasPool'] = all_features['PoolArea'].apply(lambda x: 1 if x > 0 else 0)
all_features['HasBasement'] = all_features['TotalBsmtSF'].apply(lambda x: 1 if x > 0 else 0)
all_features['Baths'] = (all_features['FullBath'] + train['BsmtFullBath'])+ 0.5*(train['HalfBath'] + train['BsmtHalfBath'])
all_features['Total_porch'] = (all_features['OpenPorchSF'] + train['3SsnPorch'] + train['EnclosedPorch'] + train['ScreenPorch'])
all_features['Exterior1st'] = all_features['Exterior1st'].fillna(all_features['Exterior1st'].mode()[0])
all_features['Exterior2nd'] = all_features['Exterior2nd'].fillna(all_features['Exterior2nd'].mode()[0])
all_features['PoolQC'] = all_features['PoolQC'].fillna('No')
all_features['Alley'] = all_features['Alley'].fillna('No')
all_features['FireplaceQu'] = all_features['FireplaceQu'].fillna('No')
all_features['Fence'] = all_features['Fence'].fillna('No')
all_features['MiscFeature'] = all_features['MiscFeature'].fillna('No')
all_features['GarageArea'] = all_features['GarageArea'].fillna(0)
all_features['GarageArea'] = all_features['GarageArea'].fillna(0)
all_features['GarageFinish'] = all_features['GarageFinish'].fillna('No')
all_features['BsmtQual'] = all_features['BsmtQual'].fillna('No')


# drop what we don't need 
all_features = all_features.drop(['GarageCond', 'BsmtFinType2', 'ExterCond', 'MasVnrArea', 'Exterior2nd', 'Condition2', 'GarageYrBlt', 'Utilities', 'Street', 'PoolQC'],axis = 1)

#function to make number and non number table 
def input_regressor(dataframe):
    numbers = dataframe.select_dtypes(include=[np.number])
    n_numbers = dataframe.select_dtypes(exclude=[np.number])
    
    cols_nan = numbers.columns[numbers.isna().any()].tolist()        
    cols_no_nan = numbers.columns.difference(cols_nan).values    

    for col in cols_nan:
        imp_test = numbers[numbers[col].isna()]   
        imp_train = numbers.dropna()         
        print('Updating Col: {}, MissingCount: {}, TrainingWith: {}' .format(col, imp_test.shape[0], imp_train.shape[0]))
        model = KNeighborsRegressor(n_neighbors=5)  
        knr = model.fit(imp_train[cols_no_nan], imp_train[col])
        numbers.loc[numbers[col].isna(), col] = knr.predict(imp_test[cols_no_nan])
    
    return pd.concat([numbers,n_numbers],axis=1)

#get input 
all_features = input_regressor(all_features)

#double make sure threre are no null values 
objects = []
for i in all_features.columns:
    if all_features[i].dtype == object:
        objects.append(i)
        if all_features[i].isnull().values.any():
            print('Updating to "No" col: {},  Values: {}' .format(i, all_features[i].isnull().sum()))
all_features.update(all_features[objects].fillna('None'))

#make new features that might be relevant 
all_features['HasWoodDeck'] = (all_features['WoodDeckSF'] > 0) * 1
all_features['HasOpenPorch'] = (all_features['OpenPorchSF'] > 0) * 1
all_features['HasEnclosedPorch'] = (all_features['EnclosedPorch'] > 0) * 1
all_features['Has3SsnPorch'] = (all_features['3SsnPorch'] > 0) * 1
all_features['HasScreenPorch'] = (all_features['ScreenPorch'] > 0) * 1
all_features['YearDif'] = all_features['YrSold'].astype(int) - all_features['YearRemodAdd'].astype(int)
all_features['HasPool'] = all_features['PoolArea'].apply(lambda x: 1 if x > 0 else 0)
all_features['Has2ndfloor'] = all_features['2ndFlrSF'].apply(lambda x: 1 if x > 0 else 0)
all_features['HasGarage'] = all_features['GarageArea'].apply(lambda x: 1 if x > 0 else 0)
all_features['HasBsmt'] = all_features['TotalBsmtSF'].apply(lambda x: 1 if x > 0 else 0)
all_features['Hasfireplace'] = all_features['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)

# gets numeric features
numeric_features = all_features.dtypes[all_features.dtypes != object].index
skewed_features = all_features[numeric_features].apply(lambda x: skew(x)).sort_values(ascending=False)
high_skew = skewed_features[skewed_features > 0.5]
skew_index = high_skew.index

# normalizes skewed features
all_feature_nskew = all_features.copy()

#log skew features
for i in skew_index:
    all_feature_nskew[i] = np.log1p(all_feature_nskew[i])

#get dummies 
all_features_dummies = pd.get_dummies(all_feature_nskew)

all_features_dummies = all_features_dummies.loc[:,~all_features_dummies.columns.duplicated()].copy()

#get all needed lists for model 
X = all_features_dummies.iloc[:len(train_label), :]
X_test = all_features_dummies.iloc[len(train_label):, :]
X.shape, train_label.shape, X_test.shape
```

## Parametric Model: 1
Lasso model

```python
# hyperparameters 
param_grid_lasso = {'alpha': [0.01, 0.1, 1, 10]}

# GridSearchCV object 
grid_lasso = GridSearchCV(estimator=Lasso(), param_grid=param_grid_lasso, cv=5)

# fit the GridSearchCV 
grid_lasso.fit(X, train_label)

# get the best hyperparameters 
best_alpha_lasso = grid_lasso.best_params_['alpha']

# train the Lasso model 
lasso = Lasso(alpha=best_alpha_lasso)
lasso.fit(X, train_label)
```

The lasso model is a linear regression model that adds a penalty term to the least squared objective function. Therefore, the assumption was that this model should perform well on the log data which was engineered earlier. As the dataset contains many columns and features, the Lasso model only trains with a smaller number of features which can prove to be successful. This model reduces the feature selection and therefore is less complex. The model was weighed with 10% in the final model blending as it did not achieve the best RMSE when comparing to other models on their own. 

### Non-parametric Model: 2
XGB Boost Model

```python
# define the hyperparameters 
param_grid_xgb = {'max_depth': [3, 5, 7],
                  'min_child_weight': [1, 3, 5],
                  'learning_rate': [0.1, 0.01, 0.001]}

# create a GridSearchCV object
grid_xgb = GridSearchCV(estimator=xgb.XGBRegressor(objective="reg:squarederror", random_state=123), param_grid=param_grid_xgb, cv=5)

# fit the GridSearchCV object 
grid_xgb.fit(X, train_label)

# get the best hyperparameters 
best_max_depth_xgb = grid_xgb.best_params_['max_depth']
best_min_child_weight_xgb = grid_xgb.best_params_['min_child_weight']
best_learning_rate_xgb = grid_xgb.best_params_['learning_rate']

# Train the XGBoost model 
xgb = xgb.XGBRegressor(max_depth=best_max_depth_xgb, min_child_weight=best_min_child_weight_xgb, learning_rate=best_learning_rate_xgb, objective="reg:squarederror", random_state=123)
xgb.fit(X, train_label)
```

The xgb model was chosen as it is a powerful gradient boosting method. The model is good in order to work with a high amount of variables which is why it was chosen for this dataset. The xgb model performed very well on the given data with a low RMSE. Therefore it has a weight of 30% in the blend of models. Moreover, this model should control overfitting as it is adding several optimization techniques. 

### Non-parametric Model: 3
Regression Forest Model

```python
#hyperparameters 
param_grid_rf = {'n_estimators': [100, 500, 1000],
                 'max_depth': [3, 5, 7],
                 'min_samples_split': [2, 5, 10]}

# create a GridSearchCV object 
grid_rf = GridSearchCV(estimator=RandomForestRegressor(random_state=123), param_grid=param_grid_rf, cv=5)

# fit the GridSearchCV object 
grid_rf.fit(X, train_label)

# get the best hyperparameters 
best_n_estimators_rf = grid_rf.best_params_['n_estimators']
best_max_depth_rf = grid_rf.best_params_['max_depth']
best_min_samples_split_rf = grid_rf.best_params_['min_samples_split']

# train Random Forest model 
rf = RandomForestRegressor(n_estimators=best_n_estimators_rf, max_depth=best_max_depth_rf, min_samples_split=best_min_samples_split_rf, random_state=123)
rf.fit(X, train_label)
```

The random forest is based on the decision tree model. Therefore, it was assumed to be suitable as it trains on multiple subsets and averages these values when finally predicting the sale price. As the xgb model, it can control overfitting which made this modle a valueable contribution to the blended model. The random forest model was weighted with 10% in the final blend as it did not perform as well as the xgb model. 

### Non-parametric Model: 4
Gradient Boosting Regressor Model

```python
#define gbr
gbr = GradientBoostingRegressor(n_estimators=3000,
                                learning_rate=0.05,
                                max_depth=4, max_features='sqrt', 
                                min_samples_leaf=15, min_samples_split=10,
                                loss='huber', random_state =42) 

#train the model 
gbr.fit(X, train_label)
```

Finally, before using the stacking method, the gbr model was run. This is another gradient boosting model which is ideal for regression models and prediciting traget values. As it is very similar to the xgb model this model is only weighted with 10% in the final blended model. However, as it is also supposed to solve overfitting and is supposed to be accurate it adds value to the final model blend.  

### Non-parametric Model: 5
Stacking Regressor Model

```python
#define stacking model
stack_gen = StackingCVRegressor(regressors=(lasso, gbr, xgb),
                                meta_regressor=xgb,
                                use_features_in_secondary=True)

#train the model
stack_gen.fit(X, train_label)
```

Lastly, the stacking method was used to optimize the results of the prior ran models. 
Stacking was used to leverage the strength of multiple models. Thefore we are using the xgb model as meta regressor which means that this model will be used to combine the before ran results. 
As regressors, lasso, gbr and xgb was used as these were well performing models. 


## Prediction of all models
Blended prediction for all the models used

```python
# all models
lasso_pred = lasso.predict(X_test)
xgb_pred = xgb.predict(X_test)
rf_pred = rf.predict(X_test)
gbr_pred = gbr.predict(X_test)
stack_gen_pred = stack_gen.predict(X_test)

# blend all models using weighted average
blend_pred = (0.1 * lasso_pred) + (0.3 * xgb_pred) + (0.1 * rf_pred) + (0.1 * gbr_pred) + (0.4 * stack_gen_pred)
               
```

In the next step, the before defined models were used to predict the data in the test set and were blended together with a weighted average to find the best possible result. As the Stacking method combines multiple models already it is weighted the highest. The xgb model was also predicting quite accurately which is why it has the second highest weight.
